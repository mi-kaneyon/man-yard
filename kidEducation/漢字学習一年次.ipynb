# 漢字学習　日本小学生1年次

```python
import pandas as pd

# Read the Excel file
file_path = 'hanzi.xlsx' # Update the path if necessary
hanzi_data = pd.read_excel(file_path)


# Function to find readings for a given Kanji (Japanese or Chinese)
def find_readings(kanji, language='JP'):
    if language == 'JP':
        readings = hanzi_data[hanzi_data['日本漢字'].str.contains(kanji, na=False)]
    else:
        readings = hanzi_data[hanzi_data['中文汉字'].str.contains(kanji, na=False) |
                             hanzi_data['繁体中文'].str.contains(kanji, na=False)]
    
    if readings.empty:
        return None
    return readings

# Function to find Kanji for a given reading (Japanese, Pinyin, or Zhuyin)
def find_kanji(reading, reading_type='JP'):
    if reading_type == 'JP':
        kanjis = hanzi_data[hanzi_data['日本よみ'].str.contains(reading, na=False)]
    elif reading_type == 'Pinyin':
        kanjis = hanzi_data[hanzi_data['中文拼音'].str.contains(reading, na=False)]
    elif reading_type == 'Zhuyin':
        kanjis = hanzi_data[hanzi_data['中文注音'].str.contains(reading, na=False)]
    
    if kanjis.empty:
        return None
    return kanjis

# Testing the functions

# Finding readings for a given Kanji (Japanese)
kanji_jp = '一'
readings_jp = find_readings(kanji_jp, language='JP')
print(f"Readings for Kanji '{kanji_jp}' (Japanese):")
display(readings_jp)

# Finding readings for a given Kanji (Chinese)
kanji_cn = '一'
readings_cn = find_readings(kanji_cn, language='CN')
print(f"\nReadings for Kanji '{kanji_cn}' (Chinese):")
display(readings_cn)

# Finding Kanji for a given reading (Japanese)
reading_jp = 'いち'
kanjis_jp = find_kanji(reading_jp, reading_type='JP')
print(f"\nKanjis for reading '{reading_jp}' (Japanese):")
display(kanjis_jp)

# Finding Kanji for a given reading (Pinyin)
reading_pinyin = 'yī'
kanjis_pinyin = find_kanji(reading_pinyin, reading_type='Pinyin')
print(f"\nKanjis for reading '{reading_pinyin}' (Pinyin):")
display(kanjis_pinyin)

# Finding Kanji for a given reading (Zhuyin)
reading_zhuyin = 'ㄧ'
kanjis_zhuyin = find_kanji(reading_zhuyin, reading_type='Zhuyin')
print(f"\nKanjis for reading '{reading_zhuyin}' (Zhuyin):")
display(kanjis_zhuyin)
```

# 出題

```python
import pandas as pd
import random
from IPython.display import display
import ipywidgets as widgets
import time

# Define columns
character_column = '日本漢字'
reading_columns = {
    'JP': '日本よみ',
    'CN': '中文拼音',
    'TW': '中文注音'
}

# Read the Excel file
file_path = 'hanzi.xlsx'
hanzi_data = pd.read_excel(file_path)
selected_questions = hanzi_data.sample(n=10)

# Global variables
question_count = 0
correct_count = 0
correct_reading = None
language_mapping = list(reading_columns.keys())

# Function to create choices
def create_choices(correct_reading, language):
    choices = [correct_reading]
    incorrect_readings = hanzi_data.sample(n=3)
    for _, row in incorrect_readings.iterrows():
        choices.append(row[reading_columns[language]])
    random.shuffle(choices)
    return choices

# Function to ask a question
def ask_question(question_number, language):
    question_row = selected_questions.iloc[question_number - 1]
    question_text_display = f"What is the reading of the character: {question_row[character_column]}?"
    correct_reading = question_row[reading_columns[language]]
    question_label.value = f"Question {question_number}"
    question_text.value = question_text_display
    choices = create_choices(correct_reading, language)
    option_buttons.options = choices
    return correct_reading

# Function to handle the answer
def handle_answer(change):
    global correct_reading, correct_count
    feedback.value = ""
    if change['new'] == correct_reading:
        correct_count += 1
    next_button.disabled = False

# Function to move to the next question
def next_question(b):
    global question_count, correct_reading
    if question_count > 0:
        feedback.value = f"Correct reading: {correct_reading} ({'Correct!' if option_buttons.value == correct_reading else 'Wrong!'})"
        time.sleep(5)  # Wait for 5 seconds
    question_count += 1
    if question_count <= 10:
        language = random.choice(language_mapping)
        correct_reading = ask_question(question_count, language)
        next_button.disabled = True
        option_buttons.value = None
    else:
        question_label.value = "Quiz Completed!"
        question_text.value = f"You have completed the quiz. Your score: {correct_count}/10"
        option_buttons.options = []
        next_button.disabled = True
        option_buttons.value = None
        
# Create widgets
question_label = widgets.Label()
question_text = widgets.Label()
option_buttons = widgets.RadioButtons()
next_button = widgets.Button(description="Next", disabled=True)
feedback = widgets.Label()

# Set up event handlers
option_buttons.observe(handle_answer, names='value')
next_button.on_click(next_question)

# Layout
quiz_box = widgets.VBox([question_label, question_text, option_buttons, next_button, feedback])

# Displaying the quiz
display(quiz_box)

# Start the quiz
next_question(None)
```
