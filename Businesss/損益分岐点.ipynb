# 損益分岐点簡易計算機

```python
import matplotlib.pyplot as plt
import numpy as np

# パラメータを設定
cost_price = 100  # 仕入れ値
selling_price = cost_price / (1 - 0.2)  # 売価
profit_margin = 0.2  # 利益率
fixed_costs = cost_price * (1 - profit_margin)  # 固定費用
variable_costs = cost_price * profit_margin  # 変動費用

# 販売数量の範囲を設定
sales = np.linspace(0, 1000, 100)

# 各販売数量に対する固定費、変動費、総費用を計算
fixed_costs_array = np.full_like(sales, fixed_costs)
variable_costs_array = variable_costs * sales
total_costs = fixed_costs_array + variable_costs_array

# 販売数量に対する売上を計算
revenue = selling_price * sales

# 損益分岐点を計算
break_even_point = fixed_costs / (selling_price - variable_costs)

# 図を描画
plt.figure(figsize=(10, 6))
plt.plot(sales, total_costs, label='Total Costs')
plt.plot(sales, revenue, label='Revenue')
plt.axvline(x=break_even_point, color='r', linestyle='--', label='Break-even Point')
plt.legend()
plt.grid(True)
plt.xlabel('Sales Quantity')
plt.ylabel('Amount')
plt.title('Break-even Analysis')
plt.show()

# 結果を表示
print(f"Break-even point: {break_even_point}")
```
