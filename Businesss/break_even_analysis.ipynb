
# 損益分岐点簡易計算機

このノートブックでは、損益分岐点を計算し、損失と利益のグラフを描画します。

以下のパラメータを使用します：
- 固定費
- 変動費
- 売上単価
- 売上数量


```python

import numpy as np
import matplotlib.pyplot as plt

# Sample parameters
fixed_cost = 30000  # Fixed cost　ここの数値を換える
variable_cost_per_unit = 50  # Variable cost per unit ここの数値を換える
selling_price_per_unit = 100  # Selling price per unit　ここの数値を換える

# Calculating break-even point
break_even_point = fixed_cost / (selling_price_per_unit - variable_cost_per_unit)

# Generating sales quantities for plotting
sales_quantities = np.arange(0, break_even_point * 2, 50)

# Calculating profits and losses based on sales quantities
profits_or_losses = [(selling_price_per_unit * q - variable_cost_per_unit * q - fixed_cost) for q in sales_quantities]

# Plotting the graph
plt.plot(sales_quantities, profits_or_losses, label="Profit/Loss")
plt.axhline(0, color='gray', linestyle='--') # Zero line for reference
plt.axvline(break_even_point, color='red', linestyle='--', label="Break-even point")
plt.xlabel('Sales Quantity')
plt.ylabel('Profit/Loss')
plt.title('Profit/Loss vs Sales Quantity')
plt.legend()
plt.show()
```
